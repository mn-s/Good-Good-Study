```python
def tosec(t):
    units = {"ns":1e-9, "us":1e-6, "ms":1e-3, "s":1}
    value, unit = t.strip().split()
    return float(value) * units[unit]

import re
info = re.findall(r"n=(.+?)\n.+?best of 3: (.+?) per loop", result.stdout)
info = [(int(t0), tosec(t1)) for t0, t1 in info]
x, y = np.r_[info].T
pl.semilogx(x, y/x, "-o");
```

```python
%%nopage
%%prun
def fib(n):
    if n < 2:
        return 1
    else:
        return fib(n-1) + fib(n-2)

def fib_fast(n, a=1, b=1):
    if n == 1:
        return b
    else:
        return fib_fast(n-1, b, a+b)

fib(20)
fib_fast(20)
```

```python
import numpy as np
from scipy.ndimage import filters
img = np.zeros((300, 300))
img[np.random.randint(0, 300, 1000), np.random.randint(0, 300, 1000)] = 255
img2 = filters.gaussian_filter(img, 4, order=2)

import io
import matplotlib
from IPython import display
buf = io.BytesIO()
matplotlib.image.imsave(buf, img2, cmap="gray")
display.Image(buf.getvalue())
```
